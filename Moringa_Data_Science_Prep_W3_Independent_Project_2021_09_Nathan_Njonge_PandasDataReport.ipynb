```python
# Load our Telcom dataset1
# CDR 20120507

url = 'http://bit.ly/Telcom_dataset1'

dataset1 = pd.read_csv(url, delimiter=',')
dataset1.head()
```

```python
# load our Telcom dataset2
# CDR 20120508

url = 'http://bit.ly/Telcom_dataset2'

dataset2 = pd.read_csv(url)
dataset2.head()
```

```python
# load our Telcom dataset3
# CDR 20120509

url = 'http://bit.ly/Telcom_dataset3'

dataset3 = pd.read_csv(url)
dataset3.head()
```

```python
# rename the Telcom dataset1 column names
dataset1.columns = ['PRODUCT','VALUE','DATE-TIME','CELL ON SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
dataset1.head()
```

```python
# rename the Telcom dataset2 column names

dataset2.columns =['PRODUCT','VALUE','DATE-TIME','CELL ON SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
dataset2.head()
```

```python
# rename the Telcom dataset3 column names

dataset3.columns =['PRODUCT','VALUE','DATE-TIME','CELL ON SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
dataset3.head()
```

```python
# Merge the three Telcom datasets
# dataset1, dataset2, dataset3

Telcom_datasets = pd.concat([dataset1,dataset2,dataset3],ignore_index=True)
Telcom_datasets.head()
```

```python
# remove duplicates from the Telcom dataset

Telcom_datasets =Telcom_datasets.drop_duplicates()
Telcom_datasets.head()
```

```python
Telcom_datasets.info()
```

```python
geo_dataset.columns
```

```python
# rename the geo dataset columns
geo_dataset.columns=['VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID','SITE_ID']
geo_dataset.head()
```

```python
# Merge the Telcom_dataset and geo_dataset
Final = Telcom_datasets.merge(geo_dataset,on='SITE_ID')
Final.head()
```

```python
Final.info()
```

```python
# Most visited City
Final.VILLES.describe()
```

```python
# City with low coverage area
Final['no per area']=1
Numbers = Final.groupby('VILLES').count().sort_values(by='AREA')
```

```python
Coverage = pd.DataFrame(Numbers['no per area'].head())
Coverage
```

```python
# city with high coverage area
Final['Number in Area']=1
Numbers =Final .groupby('VILLES').count().sort_values(by='AREA', ascending=False)
```
